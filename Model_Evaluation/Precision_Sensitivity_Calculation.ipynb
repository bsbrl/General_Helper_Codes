# Custom Functions

```python
# Function to calculate precision and sensitivity
def calculate_metrics(tp, fp, fn):
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    return precision, sensitivity
```

# Run Calculations

```python
# Define the folder path containing the .csv files
folder_path = "/Data"

# Pattern to match files containing "htcm" in their filenames
pattern = f"{folder_path}/*htcm*.csv"

# Iterate through each file matching the pattern
for file_path in glob.glob(pattern):
    # Load the CSV file
    data = pd.read_csv(file_path)

    # Extract data for calculation
    tp_head = data.loc[data['Class'] == 'True Positive', 'Head'].sum()
    tn_head = data.loc[data['Class'] == 'True Negative', 'Head'].sum()
    fp_head = data.loc[data['Class'] == 'False Positive', 'Head'].sum()
    fn_head = data.loc[data['Class'] == 'False Negative', 'Head'].sum()

    tp_tail = data.loc[data['Class'] == 'True Positive', 'Tail'].sum()
    tn_tail = data.loc[data['Class'] == 'True Negative', 'Tail'].sum()
    fp_tail = data.loc[data['Class'] == 'False Positive', 'Tail'].sum()
    fn_tail = data.loc[data['Class'] == 'False Negative', 'Tail'].sum()

    # Calculate precision and sensitivity for head and tail
    precision_head, sensitivity_head = calculate_metrics(tp_head, fp_head, fn_head)
    precision_tail, sensitivity_tail = calculate_metrics(tp_tail, fp_tail, fn_tail)

    # Print the results for each file
    print(f"Results for {file_path}:")
    print(f"Head Tracking - Precision: {precision_head*100}, Sensitivity: {sensitivity_head*100}")
    print(f"Tail Tracking - Precision: {precision_tail*100}, Sensitivity: {sensitivity_tail*100}\n")
```
